```python
# -------------------------------
# âœ… Geospatial Analysis Code
# -------------------------------

# Importing required libraries
import pandas as pd
import folium
import matplotlib.pyplot as plt
import seaborn as sns

# Load your dataset
# (Change file name if needed)
df = pd.read_csv("/content/Dataset .csv")

# -------------------------------
# ðŸ”¹ 1. Visualize restaurant locations on a map
# -------------------------------

# Remove rows with missing latitude or longitude
df = df.dropna(subset=["Latitude", "Longitude"])

# Create a base map centered at the average location
map_center = [df["Latitude"].mean(), df["Longitude"].mean()]
restaurant_map = folium.Map(location=map_center, zoom_start=10)

# Add restaurant markers to the map
for _, row in df.iterrows():
    folium.Marker(
        location=[row["Latitude"], row["Longitude"]],
        popup=f"{row['Restaurant Name']} ({row['Aggregate rating']})",
        icon=folium.Icon(color="red", icon="cutlery", prefix="fa")
    ).add_to(restaurant_map)

# Save map as HTML
restaurant_map.save("restaurant_locations_map.html")
print("âœ… Map saved as 'restaurant_locations_map.html'")

# -------------------------------
# ðŸ”¹ 2. Analyze distribution of restaurants by city
# -------------------------------

city_counts = df["City"].value_counts()

plt.figure(figsize=(10, 5))
sns.barplot(x=city_counts.index[:10], y=city_counts.values[:10])
plt.title("Top 10 Cities with Most Restaurants")
plt.xlabel("City")
plt.ylabel("Number of Restaurants")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# -------------------------------
# ðŸ”¹ 3. Correlation between location and rating
# -------------------------------

# For correlation, we can check if thereâ€™s a pattern
# between geographic coordinates and ratings

corr_lat = df["Latitude"].corr(df["Aggregate rating"])
corr_long = df["Longitude"].corr(df["Aggregate rating"])

print(f"ðŸ“ Correlation between Latitude and Rating: {corr_lat:.4f}")
print(f"ðŸ“ Correlation between Longitude and Rating: {corr_long:.4f}")

# Optional visualization
plt.figure(figsize=(8, 6))
sns.scatterplot(x="Longitude", y="Latitude", hue="Aggregate rating", data=df, palette="coolwarm", s=50)
plt.title("Restaurant Locations Colored by Ratings")
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.legend(title="Rating")
plt.show()
```
